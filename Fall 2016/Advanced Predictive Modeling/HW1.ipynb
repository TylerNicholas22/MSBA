### Part 2

The authors of the paper built a model that predicts whether a patient is about to go into cardiac arrest (i.e. Code Blue) over the next 4 hours from the time the prediction is made; a binary classification (i.e. Code Blue or not) is made at the end of each hour, resulting in 4 separate classification problems. For example, if the model is making a prediction at 8 PM, the output will be 4 separate predictions for 9PM, 10PM, 11PM and 12AM. The authors chose to use Support Vector Machines and Logistic Regression for their model; the former yielded higher recall and lower false positive rates, and the latter had more interpretable results and slightly worse accuracy.

The first step was feature extraction. Given how irregular the hospital data was (e.g. not every patient gets the same number of blood tests or check-ups, and they may receive other treatments at irregular times), the first step was to estimate certain features, such as lab test and patients' vitals, so that they are normalized by using segmented regression at various time windows. The resulting data had 381 features.

To train and assess their model, the authors used $F_{1}$ score because they cared about having both high recall and precision.
$$F_{1} = 2\frac{precision \times recall}{precision + recall}$$

Because the occurences of Code Blue is extremely rare - between 2006 and 2011, only 0.5% of the 130,000 patients had ever been flagged for Code Blue - the authors used a nested-case control method and matched Code Blue patients with non-Code Blue patients with similar demographics so that the model could learn non-demographic features that affect cardiac arrests.

Their final model predicted Code Blue with 80% recall and 20% false positive rate, compared to a recall of about 30-40% (estimated by looking at the ROC curves) at the same false positive rate from the hospital's current system.

# Question 2: Maximum likelihood estimate (10 pts)

Consider the following univariate probability distribution 
$$p(x|a,b) = \frac{x^{a-1}}{\Gamma(a)b^a}exp(-\frac{x}{b}),$$ for $x \geq 0$ and $a,b > 0.  ~~ \Gamma(a)$ is a normalization term. 

You are given N i.i.d. data points drawn from this distribution. Derive the MLE estimate for $b$, in terms of $x_1, x_2, \ldots x_N$ and $a$.

## Answer


First, we find the joint probability mass function, which can be denoated as:
$$L(a,b) = \prod_{i=1}^{N} p(x_{i}|a,b) = \prod_{i=1}^{N} = \frac{x_{i}^{a-1}}{\Gamma(a)b^a}exp(-\frac{x_{i}}{b})  $$

Next, we can simplify the expression as follows:
$$L(a,b) = \frac{(a-1)\prod_{i=1}^{N} x_{i}} {\Gamma(a)^{N}b^{N*a}} exp(\frac{\sum_{i=1}^{N}x_{i}}{-b})$$

Next, we can take the log of both sides of the function.

$$ln(L(a,b)) = ln(a-1) + ln(\prod_{i=1}^{N} x_{i}) - [ N*ln(\Gamma(a)) ] - (N*a) ln(b) - \frac{\sum_{i=1}^{N}x_{i}}{b} $$

Next, we can take the partial derivative with respect to b, and set the resulting expression equal to 0. The first 3 terms of the previous expression cancel out.
$$ 0 = \frac{N*a}{-b} + \frac{\sum_{i=1}^{N}x_{i}}{b^2}$$
$$ \frac{N*a}{b} = \frac{\sum_{i=1}^{N}x_{i}}{b^2}$$

Rearranging the terms, we get:
$$ (b^2)(N-a) = -b\sum_{i=1}^{N}x_{i}$$

Finally, we can solve for b
$$b=\frac{\sum_{i=1}^{N} x_{i}}{N*a} $$

# Question 6: Illustrating the curse of dimensionality (10 pts)

For a hypersphere of radius $a$ in $d$ dimensions, the volume is related to the surface area of a unit hypersphere (S) as
    $$ V = \frac{S \times a^d}{d}.$$


1. (6 pts) Use this result to show that the fraction of the volume which lies at values of the radius between $a - \epsilon$ and $a$, where $0 < \epsilon < a$, is given by
$f = 1 - (1 - \epsilon/a)^d$. Hence, show that for any fixed $\epsilon$, no matter how small, this fraction tends to 1 as $d \rightarrow \infty$.
2. (4 pts) Evaluate the ratio $f$ numerically, with $\epsilon/a = 0.01$, for the cases $d$ = 2, 10, and 1000.


This problem shows that if data are randomly distributed within a very high-dimensional hypersphere, most of the mass is near the surface and there is very little of it near the "core". Thus, there are few "representative" points.

## Answer

### Part 1

Let $V_{total}$ be the volume of $S$ and $V_{a-\epsilon }$ be the volume of $S$ at values of the radius between $a-\epsilon$ and $a$ such that $ \frac{V_{a-\epsilon}} {V_{total}} = f = 1 - (1 - \epsilon/a)^d. $ 

First, we can rewrite $V_{a-\epsilon }$ as:
$$\frac{V_{a-\epsilon}} {V_{total}}=\frac{\int_{a-\epsilon}^{a}Sa^{d-1}d\;a}{\int_{0}^{a}Sa^{d-1}d\;a}$$
The integral calculation is equal to:
$$\frac{V_{a-\epsilon}} {V_{total}}= \frac{\frac{S \times a^d}{d} - \frac{S(a-\epsilon)^d}{d}}{\frac{S \times a^d}{d}}$$

After simplifying the expression, we get:

$$\frac{V_{a-\epsilon}} {V_{total}} = 1- \frac{(a-\epsilon)^d}{a^d}$$ which simplifies to $$ \frac{V_{a-\epsilon}} {V_{total}}=1 - (1 - \epsilon/a)^d$$.

To show that  $f$ tends to 1 as $d \rightarrow \infty$ for any fixed $\epsilon$, we can evaluate the following:
$$\lim_{d\to\infty}1 - (1 - \epsilon/a)^d$$

Since $0 < \epsilon < a$, the possible values of $\frac{\epsilon}{a}$ are $0<\frac{\epsilon}{a}<1$. Therefore, we can derive the following:
$$\lim_{d\to\infty}1 - (1 - \epsilon/a)^d = 1- \lim_{d\to\infty}(1 - \epsilon/a)^d = 1-0 = 1 $$




### Part B

The $f$ ratios are below. As $d$ gets larger, the $f$ ratio approaches 1.

```python
def f_ratio(e_a, d):
    return 1-math.pow((1-e_a),d)

ea_ratio = .01
values = [2,10,1000]

print "F Ratios","\n"
for d in values:
    print "d={}: {}".format(d,f_ratio(ea_ratio,d))
```
